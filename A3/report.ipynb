# Q1

**Mathematical Proof can be found in the Appendix. (final pages of the report)**

In this question, I will all talk about the normalized version of magnitide of the response **(for both (1)(2)(3))**. The normalized version represents, for example, if I get some response with the parameter $\sigma$, and I will use the $\sigma^2 \times \ response$ to calculate the normalized response. We want to use convolution method on this filter to detect the blob, that is we want to split the positive part and negative part and find the summation of positive part or negative part. To be the optimal, we want to make sure the central negative area is aligned with the blob that we want to detect. However, even if we align it perfectly, magnitude of the integral of the one with larger $\sigma$ and without normalized will be smaller because there is a coefficient $\frac{1}{\sigma^2}$, which indicates the larger $\sigma$ is then the lower "response" we get. To avoid this, I will use normalised $\sigma$ so that I can compare those more meaningful.

## (1)

**Mathematical Proof can be found in the Appendix. (final pages of the report)**

$\sigma=\frac{D}{2\sqrt{2}}$. 

Consider a black circle on a white background, then the pixel value of the black circle is 0 and the white background is constant(normally 255 or 1). We want to find the maximum value in this situation since the idea is using this filter to align with the blob. Then central part of the filter is negative and the circle is 0 while the surrounding area of the filter is positive and the image is a positive constant. Then to make the maximum maganitude of response we need to find the maximum of the magnitude, we need to find the maximum value of the convolution (integral).

Based on the property of LoG filter, it is in a shape of straw hat, the central part is negative and the surrounding part is positive and the boundary is $\frac{x^2+y^2}{2\sigma^2}-1=0$. Since we want to use LoG filter to detect the black circle and we want to find the maximum magnitude response, then we need to make sure the all negative area is in the circle and no positive area is in the circle. If the center of the circle is different to the center of the filter, we cannot make all the positive be added and nothing negative be added. If there are some positive in the circle area (value 0), then we cannot get the maximum magnitude of response since we can increase the $\sigma$ to get a larger response(more positive values can be added). On the other hand, if the circle doesn't contain all negative area, then the magnitude is also not the largest since if we calculate the convolution result, there will be some negative values and we can make the $\sigma$ smaller to get rid of those negative values and get the maximum magnitude response.

Therefore, the balance point is we want to make the edge of the circle is on the division line of the filter which separate the filter into positive and negative parts. When we analysis the filter function, we notice that there are three factors in total: $\frac{1}{\pi \sigma^4}$, $\frac{x^2+y^2}{2\sigma^2}-1$ and $e^{-\frac{x^2+y^2}{2\sigma^2}}$, we can notice that the first factor and the third factor must be positive while the second factor can be both positive and negative. In this way, if we let the second term equals to 0, we can obtain a circle with radius $r=\sqrt{2}\sigma$ since $x^2+y^2=r^2 \implies \frac{r^2}{2\sigma^2}=1 \implies r^2 = 2\sigma^2 \implies r=\sqrt{2}\sigma$. Therefore $\sigma=\frac{r}{\sqrt 2}$. Since $D = 2r$, then $\sigma=\frac{D}{2\sqrt{2}}$

## (2)

**Mathematical Proof can be found in the Appendix. (final pages of the report)**

The solution to this question is same to the first question, which is $\sigma=\frac{D}{2\sqrt{2}}$. The idea of this question is similar to the first question and the difference is that in order to find the maximum magnitide of the response for the image of a white circle with a black background we are supposed to find the minimum value(most negative) of the response. 

In this way, if the filter central negative area contains the background, that means not all negative value of the filter be added to the filter. Thus, we can make it more negative by making the $\sigma$ smaller (contain more negative area will make the result more negative). On the other hand if the filter positive area contains the circle, this means the response is the summation of all negative values of the filter and also has a part of positive values, we can make the magnitude larger(do not add positive values will make the response more negative. i.e. get a larger magnitude of response) by increasing the $\sigma$ value.

Therefore, the balance point is when $x^2+y^2=2\sigma^2$ which means $\frac{D^2}{4}=2\sigma^2$. That is $\sigma=\frac{D}{2\sqrt{2}}$

```python
def build_LOG_filter(size, sigma):
    pad = size // 2
    K = np.zeros((size, size), dtype=np.float)
    for x in range(-pad, -pad + size):
        for y in range(-pad, -pad + size):
            K[y + pad, x + pad] = ((x**2+y**2)/(2*sigma*sigma)-1) * np.exp(-(x**2 + y**2)/(2*sigma*sigma))
    K /= (2 * np.pi * sigma * sigma)
#     K /= K.sum()
    return K
```

```python
def plot_filter(kernel):
    fig = plt.figure()
    ax = Axes3D(fig)
    H, W = kernel.shape
    X = np.arange(-W//2, -W//2 + W, 1)
    Y = np.arange(-H//2, -H//2 + H, 1)
    X, Y = np.meshgrid(X, Y)
    Z = kernel
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
    plt.show()
```

```python
def build_square(image_size, square_size, white_square=True):
    if white_square:
        image = np.zeros((image_size, image_size))
        image[(image_size-square_size)//2:(image_size-square_size)//2+square_size, (image_size-square_size)//2:(image_size-square_size)//2+square_size] = 1
        return image
    else:
        image = np.ones((image_size, image_size))
        image[(image_size-square_size)//2:(image_size-square_size)//2+square_size, (image_size-square_size)//2:(image_size-square_size)//2+square_size] = 0
        return image
```

```python
def conv2d(image, kernel):
    rows, cols = image.shape
    ksize, _ = kernel.shape
    padding = ksize - 1
    imgTmp = np.zeros([rows+2*padding, cols+2*padding])
    imgTmp[padding:rows+padding, padding:cols+padding] = image[:, :]
#     print(imgTmp.shape)
    output = np.zeros([rows+2*(ksize//2), cols+2*(ksize//2)])
#     print(imgTmp[31:31+ksize, 38:38+ksize] * kernel)
    
    for i in range(rows+2*(ksize//2)):
        for j in range(cols+2*(ksize//2)):
            output[i][j] = np.sum(kernel * imgTmp[i:i+ksize, j:j+ksize])
#             print(output[i][j])
    return output
```

```python
def show_response(image, kernel):
    H, W = image.shape
    result = conv2d(image, kernel)
    print(np.max(result))
    fig = plt.figure()
    ax = Axes3D(fig)
    X = np.arange(1, result.shape[1]+1, 1)
    Y = np.arange(1, result.shape[0]+1, 1)
    X, Y = np.meshgrid(X, Y)
    Z = result
#     print(result.shape)
    ax.plot_surface(X, Y, Z, cmap='rainbow')
    plt.show()
```

## (3)

Consider a given side length $s$, we want to find the value of $\sigma$ to maximises the magnitude of the response for a square of side $s$. To compare the magnitude of the response, we need to normalize our result by multiplying $\sigma^2$. The reason why we need to do the normalization is said at the beginning part of this question and the mathematical proof it can be found in the Appendix. 

```python
def show_result(length, start=0, end=0, num=1000):
    minresponse = []
    maxresponse = []
    sigma = np.linspace(start, end, num)
    for s in sigma:
        minresponse.append(np.min((s**2) * ndimage.gaussian_laplace(build_square(3*length, length, True), s, mode='constant')))
        maxresponse.append(np.max((s**2) * ndimage.gaussian_laplace(build_square(3*length, length, False), s, mode='constant')))
    
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(sigma, minresponse)
    opt = sigma[np.argmin(minresponse)]
    plt.title('White Square: Side length={}, with optimal sigma={:.4}'.format(length, opt))
    plt.xlabel('sigma')
    plt.ylabel('response')
    
    plt.subplot(1, 2,2)
    plt.plot(sigma, maxresponse)
    opt = sigma[np.argmax(maxresponse)]
    plt.title('Black Square: Side length={}, with optimal sigma={:.4}'.format(length, opt))
    plt.xlabel('sigma')
    plt.ylabel('response')
    
    plt.show()
```

```python
src_pts = np.float32([ kp1[m.queryIdx].pt for m in good_matches_without_list ]).reshape(-1,1,2)
dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good_matches_without_list ]).reshape(-1,1,2)
```

```python
def naive_ransac(src_pts, dst_pts, iters):
    assert len(src_pts) == len(dst_pts)
    length = len(src_pts)
    affine_matrix = None
    pretotal = -1
    for i in range(iters):
#         random.seed(i)
        sample_index = random.sample(range(length), 3)
        Xs = src_pts[sample_index]
        Ys = dst_pts[sample_index]
        x0 = Xs[0, 0, :]
        x1 = Xs[1, 0, :]
        x2 = Xs[2, 0, :]
        y0 = Ys[0, 0, :]
        y1 = Ys[1, 0, :]
        y2 = Ys[2, 0, :]
        pts1 = np.array([x0, x1, x2])
        pts2 = np.array([y0, y1, y2])
        M = cv2.getAffineTransform(pts1, pts2)
        inliers = 0
        for i in range(length):
            dummy_src_pt = np.concatenate([src_pts[i, 0, :], np.array([1])])
            dst_hat = np.matmul(M, dummy_src_pt)
            dst_pt = dst_pts[i, 0, :]
            if np.sum(abs(dst_hat-dst_pt)) < 5:
                inliers += 1
#         print(inliers)
        if inliers > pretotal:
#             print(inliers)
            pretotal = inliers
            affine_matrix = M
    return affine_matrix
```

```python
def show_comparison1(img1, img2, affine_matrix, src_pts, dst_pts):
    h,w = img1.shape
    pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
    img3 = img2.copy()
    img4 = img2.copy()
    
    # calculate homography matrix.
    M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
    detected_book = cv2.perspectiveTransform(pts,M)
    img3 = cv2.polylines(img3,[np.int32(detected_book)],True,255,3, cv2.LINE_AA)
    
    # calculate RANSAC matrix result.
    detected_book = np.array([np.dot(affine_matrix, np.concatenate([pts[0,0,:], np.array([1])])), 
                              np.dot(affine_matrix, np.concatenate([pts[1,0,:], np.array([1])])), 
                              np.dot(affine_matrix, np.concatenate([pts[2,0,:], np.array([1])])), 
                              np.dot(affine_matrix, np.concatenate([pts[3,0,:], np.array([1])]))]).reshape(-1, 1, 2)
    img4 = cv2.polylines(img4,[np.int32(detected_book)],True,255,3, cv2.LINE_AA)
    
    plt.figure(figsize=(12, 12))
    plt.subplot(1, 2, 1)
    plt.title('Homography Matrix')
    plt.imshow(img3, cmap='gray')
    plt.subplot(1, 2, 2)
    plt.title('Affine Transformation with RANSAC')
    plt.imshow(img4, cmap='gray')
```

Assume the proportion of inliers is $p$, which is 0.5 here, then if we want to use $n$ points to calculate the model which is 3 here. Then the probability of at least one outlier exists is $1-p^n$. Assume we need $k$ iterations to make sure the probabity of the correctness of my model is $P$, then $(1-p^n)^k$ is the probality that in there is at least one outlier for each iteration and $k$ iterations in total. Then $P=1-(1-p^n)^k$ where P is the probability that my model is correct based on $k$ iterations. Therefore, in order to find $k$, we get $1-P = (1-p^n)^k \implies ln(1-P) = kln(1-p^n) \implies k=\frac{ln(1-P)}{ln(1-p^n)}$. In current situation, $p=0.5, P=0.995, n=3$. Therefore I use $k=40$ in this part.

## (2)

In this part, I use auto update method to auto calculate the number of iterations. The newly updated iterations depends on the proportion of inliers, the expected probability of the correctness of the model we want and the number of points we used to calculate the model.

Now consider the following definitions: I define the percentage of inliers is $p$, which can be calculated by using the number of inliers in each iteration divided by the number of all matches. The expected probability is $P=0.995$ which is mentioned in the handout and the number of points that used to calculate the model $n=3$. Assumer the newly updated iteration is $k$ then $1-p^n$ is the probability that there are at least outliers when calculate the model. $(1-p^n)^k$ represents the probability that each times iteration contains at least outliers. $1-(1-p^n)^k$ is the probablity that we calculate the model with all inliers in some iteration up to k iterations, which is the probablity of the model is correct. In this way, for each time update the $k$, I use $1-P = (1-p^n)^k \implies ln(1-P) = kln(1-p^n) \implies k=\frac{ln(1-P)}{ln(1-p^n)}$

To be more specific, I will update the number of iterations when it meet two conditions:

1. The current model is better than the previous best model. That is the number of inliers calculate by the current model is the largest among all models.

2. The current proportional of inliers is greater than the one we predicate at the beginning.

However, before next step, there is an interesting point is that the number of inliers should be less than half of all matches which is indicated in the handout. I tried to use my affine model to calculate the number of inliers with really large number of iterations, what I get implies that there are only around 0.3 number of matches are inliers. There are two reasons why the handout is not accurate:

1. Half of matches are estimated by eyeballed the result, however, the threshold of inliers is less than 5 pixels, no matter we use Manhattan distance or Euler distance, 4 pixels or 6 pixels are really hard to distinguish by eyes. For example, if a point is 6 pixels to the expected point, then it will be counted as outlier by computer while it might be counted as inlier by eyes.

2. The limitation of the affine transformation. Performance of the affine tranformation is not as good as the homography transformation. Therefore, the actual inliers should be less than the inliers of homography transformation. The data on the handout might be estimated by the homography transformation and there exists some bias.

Therefore, when compared the iterations steps, I used both 0.5 and 0.3.

```python
def auto_update_iters_ransac(src_pts, dst_pts, iters):
    assert len(src_pts) == len(dst_pts)
    length = len(src_pts)
#     iters = 1
    affine_matrix = None
    pretotal = -1
    total_iters = 0
    while iters > 0:
        iters -= 1
        total_iters += 1
#         random.seed(total_iters)
        sample_index = random.sample(range(length), 3)
        Xs = src_pts[sample_index]
        Ys = dst_pts[sample_index]
        x0 = Xs[0, 0, :]
        x1 = Xs[1, 0, :]
        x2 = Xs[2, 0, :]
        y0 = Ys[0, 0, :]
        y1 = Ys[1, 0, :]
        y2 = Ys[2, 0, :]
        pts1 = np.array([x0, x1, x2])
        pts2 = np.array([y0, y1, y2])
        M = cv2.getAffineTransform(pts1, pts2)
        inliers = 0
        for i in range(length):
            dummy_src_pt = np.concatenate([src_pts[i, 0, :], np.array([1])])
            dst_hat = np.matmul(M, dummy_src_pt)
            dst_pt = dst_pts[i, 0, :]
            if np.sum(abs(dst_hat-dst_pt)) < 5:
                inliers += 1
        if inliers > pretotal:
#             print(pretotal,inliers, inliers/length)
            pretotal = inliers
            affine_matrix = M
            if inliers/length > 0.3:
                iters = math.log(1 - 0.995) / (math.log(1 - pow(inliers / length, 3)))
#                 print(total_iters,194 - total_iters, iters)
#         P = 1 - (1-(inliers/length)**3)**total_iters
#         if P > 0.995:
#             break
    print(total_iters)
    return affine_matrix
```

```python
def show_comparison2(img1, img2, src_pts, dst_pts):
    h,w = img1.shape
    pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
    img3 = img2.copy()
    img4 = img2.copy()
    
    affine_matrix1 = naive_ransac(src_pts, dst_pts, 194)
    affine_matrix2 = auto_update_iters_ransac(src_pts, dst_pts, 194)
    
    # calculate homography matrix.
    detected_book = np.array([np.dot(affine_matrix1, np.concatenate([pts[0,0,:], np.array([1])])), 
                              np.dot(affine_matrix1, np.concatenate([pts[1,0,:], np.array([1])])), 
                              np.dot(affine_matrix1, np.concatenate([pts[2,0,:], np.array([1])])), 
                              np.dot(affine_matrix1, np.concatenate([pts[3,0,:], np.array([1])]))]).reshape(-1, 1, 2)
    img3 = cv2.polylines(img3,[np.int32(detected_book)],True,255,3, cv2.LINE_AA)
    
    # calculate RANSAC matrix result.
    detected_book = np.array([np.dot(affine_matrix2, np.concatenate([pts[0,0,:], np.array([1])])), 
                              np.dot(affine_matrix2, np.concatenate([pts[1,0,:], np.array([1])])), 
                              np.dot(affine_matrix2, np.concatenate([pts[2,0,:], np.array([1])])), 
                              np.dot(affine_matrix2, np.concatenate([pts[3,0,:], np.array([1])]))]).reshape(-1, 1, 2)
    img4 = cv2.polylines(img4,[np.int32(detected_book)],True,255,3, cv2.LINE_AA)
    
    plt.figure(figsize=(12, 12))
    plt.subplot(1, 2, 1)
    plt.title('RANSAC')
    plt.imshow(img3, cmap='gray')
    plt.subplot(1, 2, 2)
    plt.title('RANSAC with auto-update iterations')
    plt.imshow(img4, cmap='gray')
```

```python
def read_images(path):
    files = os.listdir(path)
    images = []
    for file in files:
        image = mpimg.imread(path + '/' + file)
        image = rgb2gray(image)
        images.append(image)
    return np.array(images)
```

```python
def gaussian_blur(images, grad_sigma=5, time_sigma=1):
    N, H, W = images.shape
    result = []
    for n in range(N):
        blurred = ndimage.gaussian_filter(images[n,:,:], grad_sigma)
        result.append(blurred)
    result = np.array(result, dtype=np.float32)
    result = ndimage.gaussian_filter1d(result, time_sigma, axis=0)
    return result
```

```python
def lucus_kanade_one_shot(image1, image2, block_size, lmbda_thre=1e-3, lmbda_ratio_thre=100):
    H, W = image1.shape
    resultu = np.empty((H//block_size, W//block_size), dtype=np.float32)
    resultv = np.empty((H//block_size, W//block_size), dtype=np.float32)
    gradx1 = cv2.Sobel(image1, cv2.CV_64F, 1, 0, ksize=3)
    grady1 = cv2.Sobel(image1, cv2.CV_64F, 0, 1, ksize=3)
    gradx2 = cv2.Sobel(image2, cv2.CV_64F, 1, 0, ksize=3)
    grady2 = cv2.Sobel(image2, cv2.CV_64F, 0, 1, ksize=3)
    for h in range(H//block_size):
        for w in range(W//block_size):
            x, y = w*block_size, h*block_size
            Ix = gradx1[y:y+block_size, x:x+block_size]
            Iy = grady1[y:y+block_size, x:x+block_size]
            Ix2 = np.sum(Ix*Ix)
            Iy2 = np.sum(Iy*Iy)
            IxIy = np.sum(Ix*Iy)
            It = image2[y:y+block_size, x:x+block_size]-image1[y:y+block_size, x:x+block_size]
            IxIt = np.sum(Ix*It)
            IyIt = np.sum(Iy*It)
            left = np.array([[Ix2, IxIy],[IxIy, Iy2]], dtype=np.float32)
            lmbdas, _ = LA.eig(left)
            lmbda1 = lmbdas[0]
            lmbda2 = lmbdas[1]
            if lmbda1 == 0 or lmbda2 == 0 or min(lmbda1,lmbda2)<lmbda_thre or max(lmbda1, lmbda2)/min(lmbda1, lmbda2)>lmbda_ratio_thre:
                resultu[h, w] = 0
                resultv[h, w] = 0
                continue
#             print(max(lmbda1, lmbda2)/min(lmbda1, lmbda2))
            uv = np.dot(LA.inv(left), -1*np.array([IxIt, IyIt]))
#             print(h, w)
            resultu[h, w] = uv[0]
            resultv[h, w] = uv[1]
    return resultu, resultv
```

```python
def calc_gradient_orientation(gray, threshold=-1):
    gx = np.array([[-1, 0, 1], 
                   [-2, 0, 2],
                   [-1, 0, 1]], dtype=np.float32)
    gy = np.array([[-1, -2, -1],
                   [0,   0,  0], 
                   [1,   2,  1]], dtype=np.float32)
    gradx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    grady = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
#     gradx = cv2.filter2D(gray, -1, gx).astype(np.float32)
#     grady = cv2.filter2D(gray, -1, gy).astype(np.float32)
#     print(gradx[gradx<0])
#     print(grady[grady<0])
#     gradx = cv2.filter2D(build_square(50, 10), -1, gx)
#     return gradx, grady
    gradx[gradx==0] = 1e-6
#     return gradx, grady
    magnitude = np.sqrt(gradx*gradx + grady*grady)
    magnitude[magnitude<threshold] = 0
    orientation = np.arctan(grady/gradx)
    orientation = np.where(orientation<0, orientation+np.pi, orientation)
    return magnitude, orientation
```

```python
def calc_hog(mag, ori):
    assert mag.shape == ori.shape
    H = mag.shape[0] // 8
    W = mag.shape[1] // 8
    result = np.zeros((H, W, 6))
    for h in range(H):
        for w in range(W):
            row = h * 8 + (mag.shape[0]-H*8)//2
            col = w * 8 + (mag.shape[1]-W*8)//2
#             print(h, w, row, col)
            for r in range(row, row+8):
                for c in range(col, col+8):
                    result[h, w, int((ori[r, c]/(np.pi/6))+0.5)%6] += mag[r, c]
    return result
```

```python
def visualize(img, hog, tau):
    if len(img.shape) == 2:
        h, w = img.shape
    else:
        h, w, _ = img.shape
    m = h // tau
    n = w // tau
    
    theta = np.linspace(0, 2*np.pi, 16)
    r = np.linspace(0, 1, 6)
    x = np.sin(theta)[:,np.newaxis]*r
    y = np.cos(theta)[:,np.newaxis]*r
    
    X, Y = np.meshgrid(np.linspace(tau/2, (n-1)*tau+tau/2, n),
                       np.linspace(tau/2, (m-1)*tau+tau/2, m))
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(img, cmap='gray')
    for i in range(6):
        ax.quiver(X, Y,
                  np.sin(i*np.pi/6) * hog[:, :, i]*0.001,
                  np.cos(i*np.pi/6) * hog[:, :, i]*0.001,
                  color='red',
                  linewidth=0.5,
                  headlength=0,
                  headwidth=1,
                  headaxislength=0,
                  pivot='middle')
```

```python
def build_descriptors(hog):
    H, W, _ = hog.shape
    result = np.zeros((H-1, W-1, 24))
    for h in range(H-1):
        for w in range(W-1):
            result[h, w, :] = np.concatenate([hog[h, w, :], hog[h+1, w, :], hog[h, w+1, :], hog[h+1, w+1]])
            # normalize
            result[h, w, :] = result[h, w, :] / np.sum(result[h, w, :] * result[h, w, :])
    return result
```

```python
def show_result(path, threshold=-1):
    image=mpimg.imread(path)
    gray = rgb2gray(image)
    mag, ori = calc_gradient_orientation(gray, threshold)
    hog = calc_hog(mag, ori)
    visualize(image, hog, 8)
```
